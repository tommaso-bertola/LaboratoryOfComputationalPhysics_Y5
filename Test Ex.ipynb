# Ex 4.7
7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def f(n):
    m=np.arange(0,n)
    mask=np.array([True]*n)
    for i in range(2,n):
        for j in range(2,n):
            if i*j<n:
                mask[i*j]=False
    return m[mask]

%timeit f(100)
%timeit f(1000)
````

````python
def g(n):
    m=np.arange(0,n)
    mask=np.array([True]*n)
    for i in range(2, int(np.sqrt(n))):
        if mask[i]==True:
            for j in [i*(i+it) for it in range(n) if i**2+it*i<n]:
                mask[j]=False
    return m[mask]

%timeit g(100)
%timeit g(1000)
````

````python
plt.hist(m, bins=int(np.sqrt(n)))
plt.yticks([i*5 for i in range(9)],[i*5 for i in range(9)] )
plt.grid()
plt.xlabel('val')
plt.ylabel('freq')


hist, edges=np.histogram(m, bins=int(np.sqrt(n)))
centers=(edges[:-1]+edges[1:])/2
plt.errorbar(centers, hist, yerr=np.sqrt(hist),fmt='o', color='r')

print(edges[1]-edges[0])
print(np.sum(hist))
print(np.trapz(hist, dx=edges[1]-edges[0]))
plt.show()
````

````python
import scipy.stats as stats
gaussians=stats.norm(loc=m, scale=1.06*m.std()*m.size**(-1/5)).pdf(np.linspace(-2,5,300)[:,None])
plt.plot(np.linspace(-2,5,300),gaussians)
plt.show()
````

````python
conv=np.sum(gaussians, axis=1)
conv=conv/np.trapz(conv, x=np.linspace(-2,5, 300))
plt.plot(np.linspace(-2,5, 300),conv)
plt.show()
````

Imagine you have $n$ cameras looking at a spring oscillating along the $x$ axis. Each  camera record the motion of the spring looking at it along a given direction defined by the pair $(\theta_i, \phi_i)$, the angles in spherical coordinates. 

Start from the simulation of the records (say ${\cal O}(1000)$) of the spring's motion along the x axis, assuming a little random noise affects the measurements along the $y$. Rotate such dataset to emulate the records of each camera.

Perform a Principal Component Analysis on the thus obtained dataset, aiming at finding the only one coordinate that really matters.


````python
amplitude=10
omega=2
t=np.linspace(0,10,1000)
x=amplitude*np.sin(omega*t)+npr.normal(0,0.1, 1000)
y=npr.normal(0,0.1, 1000)
z=npr.normal(0,0.1, 1000)

X=np.vstack((x,y,z)).T
plt.plot(t, X[:,0], t, X[:,1], t,X[:,2])

def matrix(theta, phi):
    return np.array([[np.cos(theta)*np.sin(phi), np.cos(theta)*np.cos(phi), -np.sin(theta)*np.sin(phi)],
                     [np.sin(theta)*np.sin(phi), np.cos(theta)*np.sin(phi), np.cos(theta)*np.sin(phi)],
                     [np.cos(theta), -np.sin(phi), 0]])

XX=np.dot(matrix(10,120), X.T)
XX.shape
plt.show()
plt.plot(t, XX[0], t, XX[1], t,XX[2])
````
